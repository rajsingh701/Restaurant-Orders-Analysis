# Advance Questions

## 9. At what time are the most orders placed?

```python
ords.head(5)
```

```python
ords.dtypes
```

```python
# As we clearly see that the 'order_date' and 'order_time' are of 'object' data types. And we need to change it from object 
# to datetime format.
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
ords['order_date'] = pd.to_datetime(ords['order_date'], errors='coerce')
```

```python
ords['order_time'] = pd.to_datetime(ords['order_time'])
```

```python
ords.dtypes
```

```python
ords.head(5)
```

```python
# To find out the time of the most order placed at what time, we need to extract 'hour' from the 'order_time' column. 
```

```python
ords['hour'] = ords['order_time'].dt.hour
```

```python
# Now count the values by hours 
```

```python
hourly_orders = ords['hour'].value_counts().sort_index()
```

```python
peak_hour = hourly_orders.idxmax()
max_orders = hourly_orders.max()
```

```python
print('Peak Hour:', peak_hour)
print('Orders:', max_orders)
```

## 11. What is the relationship between item price and order frequency?

```python
rev.head(5)
```

```python
# frequency order
feq_ord = rev['item_name'].value_counts().rename('order_count')
```

```python
# price per item 
price = rev.drop_duplicates('item_name').set_index('item_name')['price']
```

```python
#combine 
analysis = pd.concat([feq_ord, price], axis=1).dropna()
```

```python
analysis.sort_values('order_count', ascending=False)
```
